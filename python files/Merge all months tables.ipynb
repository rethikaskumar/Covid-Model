```python
import pandas as pd
```

```python
listdf=[]
```

```python
cols={0:'District', 1:'TotalPositive', 2:'Discharged', 3:'ActiveCases', 4:'Death', 5:'Date'}
```

```python
for month in ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November',
              'December']:
    print(month)
    df=pd.read_csv(month+'.csv', header=None)
    listdf.append(df)
```

```python
maindf=pd.concat(listdf)
```

```python
maindf=maindf.rename(columns=cols)
```

```python
maindf.to_csv('final.csv', index=False)
```
