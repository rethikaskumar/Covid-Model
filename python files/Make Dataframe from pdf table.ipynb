```python
import os
```

```python
import re
```

```python
import pandas as pd
```

```python
from PyPDF2 import PdfReader
```

```python
cols={'index':'District', 0:'TotalPositive', 1:'Discharged', 2:'ActiveCases', 3:'Death'}
```

```python
Mon=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
maindf=pd.DataFrame()
```

```python
mainfile=open('covid.csv', 'a')
```

```python
for month in Mon:
    path='D:/interns/Covid data/'+month
    fileList=os.listdir(path)
    monthfile=open(month+'.csv', 'w')
    monthdf.append(maindf)
    for f in fileList:
        print(f)
        p=path+'/'+f
        file=PdfReader(p)
        nump=len(file.pages)
        text=""
        for i in range(nump):
            if file.pages[i] is not None:
                #if f=='Media-Bulletin-29-05-21-COVID-19.pdf':
                    #print(file.pages[i])
                text=text+file.pages[i].extract_text()
        text=re.split(r'District Wise Case Details', text)[1]
        text=(re.split(r'/*Note', text)[0])
        text=text.replace("  ", " ")
        text=re.sub(r'(\w\s)\n(\()', '\g<1>\g<2>', text)
        text=re.sub(r'([a-z])\s([A-Z]|\()', '\\1\\2', text)
        text=re.sub(r'([0-9]),([0-9])', '\\1\\2', text)
        list2=text.split('\n')
        date=list2[0][4:-3]
        dct={}
        for i in list2[5:-2]:
            i=i.split(' ')
            if len(i)==7:
                monthfile.write(','.join(i[1:])+date+'\n')
    monthfile.close()
```
